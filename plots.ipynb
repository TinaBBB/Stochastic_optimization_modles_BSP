```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
results_realCap = pd.read_csv('./results/results_EF.csv')
results_fixedCap = pd.read_csv('./results/results_EF_fixedCap.csv')
```

```python
distributions = ['log-normal', 'normal', 'uniform', 'exponential']
x = [100,300,500,700,900,1000]


#y = results_realCap[(results_realCap['Scenario Number'] == 100) & (results_realCap['Distribution'] == 'log-normal')]
rt_lognormal = list(results_realCap[results_realCap['Distribution'] == 'log-normal'].duration)
rt_normal = list(results_realCap[results_realCap['Distribution'] == 'normal'].duration)
rt_exponential = list(results_realCap[results_realCap['Distribution'] == 'exponential'].duration)
rt_uniform = list(results_realCap[results_realCap['Distribution'] == 'uniform'].duration)

plt.xlabel('scenari number') 
plt.ylabel('runtime (sec)')

lognormal_ln, = plt.plot(x,rt_lognormal, marker='o', color='green', label='log-normal')
normal_ln, = plt.plot(x,rt_normal, marker='*', color='blue', label='normal')
exponential_ln, = plt.plot(x,rt_exponential, marker='D', color='orange', label='exponential')
uniform_ln, = plt.plot(x,rt_uniform, marker='H', color='magenta', label='uniform')
plt.legend(handles=[lognormal_ln, normal_ln,exponential_ln, uniform_ln])

plt.show()
```
